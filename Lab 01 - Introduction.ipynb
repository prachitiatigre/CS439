### Linearly Spaced Points
There are many tasks that you need to generate a set of points in a range that are linearly spaced. For this task
we will use numpy modules arnage and linspace.

Note that the tutorial uses `np.arange`. While this is fine in some cases, we generally prefer to use `np.linspace`. `np.linspace(a, b, N)` divides the interval `[a, b]` into N points, while `np.arange(a, b, s)` will step from `a` to `b` with a fixed step size `s`.

One thing to keep in mind is that `np.linspace` always includes both end points while `np.arange` will *not* include the second end point `b`. For this reason, when we are plotting ranges of values we tend to prefer `np.linspace`.

Write two functions that return the same result.

```python
import numpy as np

# create an array of linearly spaced points from -5(inclusive) to 6 (not inclusive) spaced by 1
A = np.arange(-5,6,1)
print(A)

# do the same with linspace
B = np.linspace(-5, 5, 11, dtype = 'int')
print(B)
```

```python
### BEGIN SOLUTION
import numpy as np
import matplotlib.pyplot as plt

x = [2, 8]
y = [2, 8]
fig = plt.figure(figsize=(8,8))
t = np.linspace(0, np.pi, 500)
counter = 1
plt.suptitle("Sine waves with varying a=[2,8], f=[2,8]")

for f in y:
    for a in x:
        plt.subplot(2, 2, counter)
        plt.plot(t, a*np.sin(2*np.pi*f*t))
        plt.xticks([0, np.pi/2, np.pi], ['0', 'π/2', 'π'])
        plt.axis([0, (np.pi),-10, 10])
        plt.xlabel('a: ' +str(a))
        plt.ylabel('f: ' +str(f))
        counter = counter + 1
        plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)

plt.show()
        
### END SOLUTION
```

The code below plot function f(x,y) in 3D. The function is 
$$f(x,y) = x^2 + y^2 + 1$$
### Activity 1

### Activity 2
Change the function to $$f(x,y) = xy $$ and re-plot. Estimate the values (x,y) that minimizes f(x,y)?

Answer: The values of (x,y) that minimizes the function f(x,y) are (6,-6) and (-6,6)

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    # return (x**2 + y**2)
    return x*y

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

## Task 3 - Reading Documentation
We should also familiarize ourselves with looking up documentation and learning how to read it. Below is a section of code that plots a basic wireframe. Replace each `# Your answer here` with a description of what the line below does, what the arguments being passed in are, and how the arguments are used in the function. For example your answer may look like,

`np.arange(2, 5, 0.2)`

`# This returns an array of numbers from 2 to 5 with an interval size of 0.2`

Hint: The `Shift` + `Tab` (to see what a function does) tip from earlier in the notebook may help here.

```python
from mpl_toolkits.mplot3d import axes3d

# The linspace() method returns evenly spaced numbers over a specified interval.
# The following returns an array of 100 numbers from 1.5π to -1.5π that are evenly spaced.
# The datatype of the numbers that are produced is float.
u = np.linspace(1.5*np.pi, -1.5*np.pi, 100)

# Returns coordinate matrices from coordinate vectors.
# u is a 1 dimensional array that represents the coordinates of the grid.
# x is the row of the matrix and y is the column of the matrix.
# x will contain elements of u and y will also contain elements of u since the two parameters
# of meshgrid are u and u.
[x,y] = np.meshgrid(u, u)

# The flatten() method returns a copy of the array collapsed into a one-dimension array.
# The sqrt() method returns the square root of a value.
# In this case, the variable 'squared' stores the square root of the sum of the square of 
# x.flatten() and the square of y.flatten().
# The cos() method returns the cosine value of all the elements of 'squared' and stores that in z.
squared = np.sqrt(x.flatten()**2 + y.flatten()**2)
z = np.cos(squared)

# The reshape() method changes the shape of the array without changing its data. 
# Here it is changing the dimension of z into the dimension of x
z = z.reshape(x.shape)

# The figure() method creates a new figure
# The add_subplot() method creates a 3 axis object which is specified by the parameter projection="3d"
# The variable 'ax' is a variable that can be used to add any plot to the figure that is created
# The '111' indicates the row, column and index position of the graph.
# Here the row, column and index position are 1
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# The plot_wireframe() method plots a 3D wireframe.
# x, y and z are 2D arrays.
# The rstride parameter is the stride of the row direction and it is to 10.
# The cstride parameter is the stride of the column direction and it is to 10.
ax.plot_wireframe(x, y, z, rstride=10, cstride=10)

# The view_init() method can be used to rotate the axes of the graph programmatically.
# 'elev' is the elevation angle in the z plane and 'azim' is the azimuth angle in the x,y plane
# Here the elevation angle is 50 degrees and the azimuth angle is 30 degrees
ax.view_init(elev=50., azim=30)

# This saves the image that is created after plotting the graph, to your computer
# Here it saves as figure1.png on your computer
plt.savefig("figure1.png")
```
